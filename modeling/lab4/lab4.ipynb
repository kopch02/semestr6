```python
import random

P_A = 0.8
P_B = 0.7
P_C = 0.95
P_D = 0.85
P_E = 0.9
P_F = 0.7

N = 100


def ne(a):
    return 1 - a


def bloc_1(a, b):
    a_is_work = False
    b_is_work = False
    if a <= P_A:
        a_is_work = True
    if b <= P_B:
        b_is_work = True
    return a_is_work | b_is_work


def bloc_2(c):
    c_is_work = False
    if c <= P_C:
        c_is_work = True
    return c_is_work


def bloc_3(d, e, f):
    d_is_work = False
    e_is_work = False
    f_is_work = False
    if d <= P_D:
        d_is_work = True
    if e <= P_E:
        e_is_work = True
    if f <= P_F:
        f_is_work = True
    return d_is_work | e_is_work | f_is_work


def system(a, b, c, d, e, f):
    return bloc_1(a, b) & bloc_2(c) & bloc_3(d, e, f)


count_ok = 0
count_error = 0
p_bloc_1 = 1 - ne(P_A) * ne(P_B)
p_bloc_2 = 1 - ne(P_C)
p_bloc_3 = 1 - ne(P_D) * ne(P_E) * ne(P_F)
p_teor = p_bloc_1 * p_bloc_2 * p_bloc_3


for i in range(N):
    a = random.random()
    b = random.random()
    c = random.random()
    d = random.random()
    e = random.random()
    f = random.random()

    if system(a, b, c, d, e, f):
        count_ok += 1
    else:
        count_error += 1

p = count_ok / N
error = abs(p - p_teor)

print(f'Полученная вероятность: {p}')
print(f'Теоретическая вероятность: {p_teor}')
print(f'Погрешность: {error}')
```

```python
p_error_alu = 0.2
p_error_ram = 0.3
p_error_other = 0.5

p_detect_alu = 0.8
p_detect_ram = 0.9
p_detect_other = 0.95

p_detect_teor = p_error_alu * p_detect_alu + \
                p_error_ram * p_detect_ram + \
                p_error_other * p_detect_other

N = 100

def detect_error():
    p = random.random()
    if p < p_error_alu:
        if random.random() < p_detect_alu:
            return 1
        else:
            return 0
    elif p < p_error_alu + p_error_ram:
        if random.random() < p_detect_ram:
            return 1
        else:
            return 0
    else:
        if random.random() < p_detect_other:
            return 1
        else:
            return 0

count_detected_errors = 0
for i in range(100):
    count_detected_errors += detect_error()

print(f'Вероятность обнаружения ошибки: {count_detected_errors / N}')
print(f'Теоретическая вероятность обнаружения ошибки: {p_detect_teor}')
print(f'Погрешность вероятности обнаружения ошибки: {abs(p_detect_teor - count_detected_errors / N)}')
```

```python
x_min = -2
x_max = 2
y_min = -2
y_max = 2

s_teor = 6.84359


def is_hit(x, y):
    state = False
    if -x**2 + y**3 < -1 and x + y < 1:
        state = True
    return state


def get_square(n):
    count = 0
    for i in range(n):
        x = random.random() * (x_max - x_min) + x_min
        y = random.random() * (y_max - y_min) + y_min

        if is_hit(x, y):
            count += 1
    s0 = (x_max - x_min) * (y_max - y_min)
    s = s0 * count / n
    return s


n = 10
while n <= 100000:
    print(
        f'Площадь фигуры при n = {n} : {get_square(n)}, погрешность: {abs(s_teor - get_square(n))}'
    )
    n *= 10
```

```python
import numpy as np
import pandas as pd


def number_4():
    N = 1000
    demands = [10000, 20000, 40000, 60000]
    probs = [0.1, 0.35, 0.3, 0.25]
    profit_arr = np.zeros(len(demands))

    for i in range(len(demands)):
        for _ in range(N):
            # случайное вещ число из равномерного распределения
            p = np.random.uniform()
            if (0 <= p < 0.1):
                demand_t = demands[0]
            elif (0.1 <= p < 0.45):
                demand_t = demands[1]
            elif (0.45 <= p <= 0.75):
                demand_t = demands[2]
            elif (0.75 <= p < 1):
                demand_t = demands[3]
            profit = 0
            if (demands[i] == 10000):
                profit = 10000 * 2.5
            if (demands[i] == 20000):
                if (demand_t == 10000):
                    profit = 10000 * 2.5 - 1.3 * 10000
                else:
                    profit = 20000 * 2.5
            if (demands[i] == 40000):
                if (demand_t == 10000):
                    profit = 10000 * 2.5 - 30000 * 1.3
                elif (demand_t == 20000):
                    profit = 20000 * 2.5 - 1.3 * 20000
                else:
                    profit = 40000 * 2.5
            if (demands[i] == 60000):
                if (demand_t == 10000):
                    profit = 10000 * 2.5 - 50000 * 1.3
                elif (demand_t == 20000):
                    profit = 20000 * 2.5 - 1.3 * 40000
                elif (demand_t == 40000):
                    profit = 40000 * 2.5 - 1.3 * 20000
                else:
                    profit = 60000 * 2.5
            profit_arr[i] += profit

        profit_arr[i] /= N
    df = pd.DataFrame([demands, probs, profit_arr],
                      index=["Спрос", "Вероятность", "Средняя прибыль"]).T
    index = df["Средняя прибыль"].idxmax(0)
    print(df)
    print(f"Наименее рискованное решение находится под индексом: {index}")


number_4()
```

```python
import random

n = 100000
detail_cost = 35
material_cost = 20
expenditure_cnc = 5
expenditure_polish = 4
expenditure_gt_thick = 3
expenditure_polish_correction = 2
cnc_less = 0.08
cnc_gt = 0.12
polish_up_proc = 0.03
polish_dn_proc = 0.06
cost = 0
detail_count = 0
profit = 0
for _ in range(n):
    cost += material_cost + expenditure_cnc
    r = random.uniform(0, 1)
    if r < cnc_less: continue
    if r > cnc_less and r < cnc_gt + cnc_less: cost += expenditure_gt_thick
    cost += expenditure_polish
    r1, r2 = random.uniform(0, 1), random.uniform(0, 1)
    if r1 < polish_up_proc and r2 < polish_dn_proc: continue
    if (r1 < polish_up_proc) ^ (r2 < polish_dn_proc):
        cost += expenditure_polish_correction
    profit += detail_cost
    detail_count += 1
print(f"Вероятность выпуска годной детали: {detail_count / n}")
print(f"Средняя прибыль предприятия: {(profit - cost) / n}")
```
